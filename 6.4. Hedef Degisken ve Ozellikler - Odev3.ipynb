```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format

baslik_font = {'family': 'arial','color':  'darkred','weight': 'bold','size': 15 }
eksen_font = {'family': 'arial','color':  'darkblue','weight': 'bold','size': 10 }
```

```python
# Ev fiyatları modelinizi tekrar çalıştırın ve sonuçları yorumlayın. Hangi özellikler istatistiksel olarak anlamlı 
# ve hangileri değildir?
ev = pd.read_csv('house_prices_train.csv')
```

```python
# Bir onceki odevde yaptigimdan farkli olarak korelasyon martisinde ev fiyatlariyla korelasyonu 0.5 uzeri olan 
# degiskenleri modelim icin tutacagim

Y = ev['SalePrice']
X = ev[['OverallQual','GrLivArea','GarageArea','GarageCars','YearBuilt','YearRemodAdd','TotalBsmtSF','1stFlrSF',
       'FullBath','TotRmsAbvGrd']]
```

```python
X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```

```python
# GarageArea ve TotRmsAbvGrd (Total rooms above grade) yuksek P degeri verdikleri icin anlamli gozukmuyorlar.
```

```python
# Şimdi, anlamsız özellikleri modelinizden hariç tutun. Bir şey değişti mi?
Y = ev['SalePrice']
X = ev[['OverallQual','GrLivArea','GarageCars','YearBuilt','YearRemodAdd','TotalBsmtSF','1stFlrSF','FullBath']]
```

```python
X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```
