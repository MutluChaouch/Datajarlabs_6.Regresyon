## 6.5. Modelin Egitim Performansini Olcme - Odev2

```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format

baslik_font = {'family': 'arial','color':  'darkred','weight': 'bold','size': 15 }
eksen_font = {'family': 'arial','color':  'darkblue','weight': 'bold','size': 10 }
```

```python
ev = pd.read_csv('house_prices_train.csv')
```

```python
# Bir onceki odevde sectigim degiskenleri modelim icin tutacagim.
# Ev fiyatları modelinizi tekrar çalıştırın ve F-testi, R-kare, ayarlanmış R-kare, AIC ve BIC kullanarak 
# modelinizin uygunluğunu değerlendirin. Modelinizin tatmin edici olduğunu düşünüyor musunuz? Açıklayın

Y = ev['SalePrice']
X = ev[['OverallQual','GrLivArea','GarageArea','GarageCars','YearBuilt','YearRemodAdd','TotalBsmtSF','1stFlrSF',
       'FullBath','TotRmsAbvGrd']]
X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```

```python
# 1. model:
R-kare: 0.774
Ayarlanmis R-kare: 0.772
F-statistics: 495.4
AIC: 34490
BIC: 35000
    
# Modelin R-kare ve Ayarlanmis R-kare degerleri yuksek olmasina ragmen, daha yuksek cikan bir model bulabiliriz
# diye durunuyorum. Bazi degiskenlerin P degerleri de zaten yuksek cikti.
```

```python
# Modelinizin uygunluğunu iyileştirmek için, bazı değişkenleri ekleyerek veya kaldırarak farklı model özelliklerini
# deneyin.
# P degeri yuksek olan degiskenleri modelden cikararak deneyecegim.
```

```python
Y = ev['SalePrice']
X = ev[['OverallQual','GrLivArea','GarageCars','YearBuilt','YearRemodAdd','TotalBsmtSF','1stFlrSF']]
X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```

```python
# Denediğiniz her model için, uygun metrikleri alın ve modellerinizi birbiriyle karşılaştırın. 
# Hangi model en iyisidir ve neden?

# Ikinci modelde bazi degiskenleri cikararak daha iyi bir sonuc elde edemedim. 
# Kalite ve Renovasyon yilinin onemli oldugunu dusundugum icin bu ikisinin etkilesimini de ekleyerek bakacagim.

Y = ev['SalePrice']
ev['qual_yearremod'] = ev.OverallQual * ev.YearRemodAdd
X = ev[['OverallQual','GrLivArea','GarageCars','YearBuilt','YearRemodAdd','TotalBsmtSF','1stFlrSF','qual_yearremod']]
X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```

```python
# 1. model:
R-kare: 0.774
Ayarlanmis R-kare: 0.772
F-statistics: 495.4
AIC: 34490
BIC: 35000
    
# 3. model:
R-kare: 0.784
Ayarlanmis R-kare: 0.783
F-statistics: 660.1
AIC: 34870
BIC: 34910
    
# dusuk bir fark olmasina ragmen ucuncu yani kalitenin renovasyon yili ile etkilesimli oldugu modelin aciklayiciligi
# 1. ve 2. modele gore daha fazladir.
```
